```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel('dataframe3.xlsx')
data_feature = data.iloc[:,13:]
data_feature['cumVowel'] = data_feature['cumVowel'] / 3
data_feature['vowel'] = data_feature['vowel'] / 3
data_feature['double'] = data_feature['double'] / 2
data_feature.to_excel('data_feature.xlsx', index=False)
data_feature
```

```python
data_feature.drop(['double','vowel','cumVowel','uncommonwords'], axis=1, inplace=True)
data_feature
```

```python
data_label = data.iloc[:, 6:13]
data_label
```

```python
data_feature = data_feature.values
data_label = data_label.values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_feature, data_label, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100, random_state=0)
regressor.fit(X_train, y_train)
regressor.score(X_test, y_test)
```

```python
X = np.array([0.8579387186629527, 1.2000000000000002, 2.0])
result = regressor.predict(X.reshape(1, -1))
result, np.sum(result)
```
